```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://www.premierleague.com/tables"
r = requests.get(url)  # r stands for response
print(r)
```

```python
# getting information and parse it with BeautifulSOup

soup = BeautifulSoup(r.text, 'html.parser')
print(soup)
```

```python
league_table = table.find_all("th")
league_table
```

```python
# Creating column Header
import re   # to handle duplciates

headers = []

for item in league_table:
    text = item.text.strip()
    text = re.sub(r'\b(\w+)\b\s+\1+', r'\1', text)
    text = text.split("\n")[0]
    headers.append(text)

print(headers)
```

### Extracting data and push to df

```python
rows = table.find_all("tr")
data = []

for row in rows:
    cols = row.find_all("td")
    cols = [ele.text.strip() for ele in cols]
    data.append(cols)

# Create a DataFrame
df = pd.DataFrame(data, columns=headers)
df
```

### Cleaning the data

```python
#checking the data scrape form the website
df
```

```python
# Since theres is data that is duplicated and unwanted, dropping the row by subset column method
df = df.dropna(subset=["Club"])
df.head()
```

```python
# Resetting the index
df.reset_index(drop=True, inplace=True)
df.head(10)
```

```python
# inspecting  and cleaning column 'Position'
df["Position"]
```

```python
df["Position"] = df["Position"].str.split("\n\n\n\n").str[0]
df.head()
```

```python
# inspecting  and cleaning column 'Club'
df["Club"]
```

```python
df["Club"] = df["Club"].str.split("\n").str[0]
df.head()
```

```python
# Selecting needed columns only for further analysis
#df.columns

df = df[['Position', 'Club', 'Played', 'Won', 'Drawn', 'Lost', 'GF', 'GA', 'GD',
       'Points']]
df.head()
```

```python
# saving df to csv
df.to_csv("Premiere_League_Table_2024.csv", header=True, index=False)
```
