```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://vintagesac.com/on-sale/"
```

```python
response = requests.get(url)
page = response.content
```

```python
soup = BeautifulSoup(page, "html.parser")
print(soup)
```

```python
items = soup.find_all("div", class_="woocommerce")
items
```
