```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Palm_oil_production_in_Malaysia"
r = requests.get(url)
#print(r)
```

```python
soup = BeautifulSoup(r.text, "html.parser")
print(soup)
```

```python
# finding table we want
table = soup.find_all("table")[1]
print(table)
```

```python
# extracting the header
palm = table.find_all("th")
print(palm)
```

```python
# Creating column header

headers = []

for item in palm:
    text = item.text.strip()
    headers.append(text)
    
headers = headers[1:] 
#headers

# transform into df
df = pd.DataFrame(columns = headers)
df
```

```python
# since the table has one unwanted header, reselecting column is needed to make sure column header match the column data

#df.columns

df = df[['State','Plantation area (hectares)', 'Percentage of total plantation area','Percentage of state area']]
df
```

### Extracting table data 'td' from table row 'tr'



```python
rows = table.find_all("tr")
#print(rows)

row_data = []

for row in rows:
    data = row.find_all("td")
    data = [element.text.strip() for element in data]
    row_data.append(data)

#print(row_data)

# extract to df

df = pd.DataFrame(row_data, columns = headers)
df
```

### cleaning the data


```python
#removing unwanted rows
palm_area = df.drop([0,1])
palm_area = palm_area.reset_index(drop = True)
palm_area.head()
```

```python
palm_area.to_csv("palm_plantation_area_2016.csv", header = True, index= False)
```
