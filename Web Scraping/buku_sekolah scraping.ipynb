```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://badanbookstore.com/product.php?mgroupid=Malay+Books&sgroupid=Buku+Teks&subcatid=KSSM+TING.+5"
```

```python
response = requests.get(url)
#print(response)
```

```python
page= response.content
#print(page)
```

```python
soup = BeautifulSoup(page, "html.parser")
print(soup)
```

```python
# the issue experienceing here is, the web page is made form tables inisde tables, so i have to find specific table that i need to attack
# since there is no specific classes to attack the tale that i wanna scrape. i have to find manually the index of table

#table = len(soup.find_all("table"))     --> 196 table -- 172
tables = soup.find_all("table")[172:193]
tables
```

```python
books = []

for table in tables:
    title = table.find("span", class_="textHIGHLIGHT").text
    #print(title)
    price = table.find("span", class_="textFAQ").text
    #print(price)
    author = table.find("span", class_="textBOOKAUTHUR").text
    #print(author)
    books.append([title, price, author])

print(books)
```

```python
df = pd.DataFrame(books, columns = ["Title","Price","Author"])
df.to_csv("buku_sekolah_tingkatan_5.csv", index= False)
```
