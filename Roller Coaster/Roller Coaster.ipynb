```python
df = df.rename(columns = 
          {'Location':'location', 'Status':'status',
            'Manufacturer':'manufacturer', 'Type_Main':'type_main', 
           'Inversions_clean':'inversion_clean', 'Gforce_clean':'gforce_clean'
          }
          )
          
```

```python
# checking base one identical row name if the data is really duplicated   --> the data is duplicated

df.query('coaster_name == "Crystal Beach Cyclone"')
```

```python
# checking multiple duplicated data
# reinverse the data and reassign to avoid duplicated data

df = df.loc[~df.duplicated(subset = ['coaster_name','location','opening_date_clean'])]
```

```python
df = df.reset_index(drop = True).copy()
df
```

```python
#Univariate Analysis   --- Analysis by comparing usin one variable only

df['year_introduced'].value_counts()
```

```python
# saving chart as matplotlib axis
ax = df['year_introduced'].value_counts().head(11).plot(kind ='bar', 
                                                        title = 'Top 10 Years Coasters Intriduced', 
                                                        legend = True, 
                                                        color = 'magenta')   #save as ax too ease the maipulation of the chart
ax.set_xlabel('year_introduced')     
ax.set_ylabel('Count')
```

```python
#histogram plot
df['speed_mph'].plot(kind = 'hist', 
                     title = 'Coasters Speed(mph)', 
                     xlabel = 'Speed (mph)', 
                     color = 'beige', 
                     bins = 20, 
                     alpha = 0.9)
```

```python
# Density Plot
df['speed_mph'].plot(kind = 'kde', 
                     title = 'Coasters Speed(mph)', 
                     xlabel = 'Speed (mph)', 
                     color = 'red', 
                     alpha = 0.9)
```

```python
# Bivariate Analysis
# scatterplot

df.plot(kind = 'scatter',
        x = 'speed_mph', xlabel = 'Height (ft)',
        y = 'height_ft', ylabel = 'Speed (mph)',
        title = 'Coaster Speed Vs Height',
        color = 'green',
        s = 50,
        grid = True)
plt.show()   #cleaner way to show the chart
```

```python
df.head()
```

```python
# using seaborn for the chart
sns.scatterplot(x = 'speed_mph', 
                y = 'height_ft',
                data = df,
                hue = 'year_introduced',
                s = 100,
                palette = 'coolwarm'
                )
plt.title('Coaster Speed Vs Height')
plt.show()
```

```python
# Multivariate Analysis

sns.pairplot(df,vars = ['year_introduced','speed_mph','height_ft','inversion_clean','gforce_clean'], hue = 'type_main', palette = 'coolwarm')
plt.show()
```
