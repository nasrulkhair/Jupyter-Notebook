##### 1.Finding the null value and replacing the null with mean value

```python
# finding the location of the null values
null_loc = df.isna()
for column in null_loc:
    null_indices = null_loc[column].index[null_loc[column]]
    if not null_indices.empty:
        print(f"Null values in column '{column}': {null_indices.tolist()}")
```
