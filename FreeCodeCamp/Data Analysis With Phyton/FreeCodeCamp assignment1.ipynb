```python
import pandas as pd
```

```python
# read data from file
df = pd.read_csv('adult.data.csv', index_col=False, header=None)
df
```

```python
# adding column names
df.col_names = ["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","salary"]
df.columns = df.col_names
df.head()
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

race_count = df["race"].value_counts()
race_count
```

```python
df["sex"].unique()
```

```python
# striping the blank space

df["sex"] = df["sex"].str.strip(" ")
```

```python
 # What is the average age of men?

average_age_men = df[df["sex"] == " Male"]["age"].mean()
average_age_men
```

```python
df["education"].unique()
```

```python
df["education"] = df["education"].str.strip(" ")
df["education"].unique()
```

```python
# What is the percentage of people who have a Bachelor's degree?

total_count = len(df["education"])
total_count

bachelor_count = (df["education"] == "Bachelors").sum()
bachelor_count

percentage_bachelors = round((bachelor_count / total_count) * 100, 2)
percentage_bachelors
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

df["salary"] = df["salary"].str.strip(" ")
df["salary"].unique()

higher_education = df[df["education"].isin(["Bachelors","Masters","Doctorate"])]
higher_education

higher_education_rich = (higher_education["salary"] == ">50K").sum()
higher_education_rich

percentage_higher_education_rich = round((higher_education_rich / len(df["salary"])) * 100, 2)
percentage_higher_education_rich
```

```python
# What percentage of people without advanced education make more than 50K?

lower_education = df[~df["education"].isin(["Bachelors","Masters","Doctorate"])]
lower_education

lower_education_rich = (lower_education["salary"] == ">50K").sum()
lower_education_rich

percentage_lower_education_rich = round((lower_education_rich / len(df["salary"])) * 100, 2)
percentage_lower_education_rich
```

```python
df.head()
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?

min_work_hours = df["hours_per_week"].min()
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?

min_work_hours = df[df["hours_per_week"]==1]
min_work_hours

min_work_hours_rich = (min_work_hours["salary"] == ">50K").sum()
min_work_hours_rich

rich_percentage = round((min_work_hours_rich / len(df["hours_per_week"])) * 100, 4)
rich_percentage
```

```python
# What country has the highest percentage of people that earn >50K?

salary2 = df[df["salary"] == ">50K"]
salary2
```

```python
highest_earning_country = salary2["native_country"].value_counts().idxmax()
highest_earning_country
```

```python
highest_earning_country_percentage = round((salary2["native_country"].value_counts() / df["native_country"].value_counts()) * 100, 2).idxmax()
highest_earning_country_percentage
```

```python
 # Identify the most popular occupation for those who earn >50K in India.

salary2

occupation_salary2 = salary2["occupation"]
occupation_salary2

top_IN_occupation = occupation_salary2.groupby(occupation_salary2).size().idxmax()
top_IN_occupation
```
