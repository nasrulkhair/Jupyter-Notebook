```python
import requests

def fetch_crude_oil_data(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    # Make the request with custom headers
    response = requests.get(url, headers=headers)

    # Check if the request was success
    if response.status_code == 200:
        #Read the HTML content into a df
        dfs = pd.read_html(response.text)
        
        # display the first df
        if dfs:
            df = dfs[0]
            return df
        else:
            print("No tables found")
            return None
    
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

# URL of the webpage containing the crude oil price history chart
url = 'https://www.macrotrends.net/1369/crude-oil-price-history-chart'

#Fetch the data
crude_oil_data = fetch_crude_oil_data(url)

# Display the data if available
if crude_oil_data is not None:
    print(crude_oil_data)
    
```

##### There is significant spike in crude oil price. Lets check what happen to that give such an impact to the crude oil price?
The significant spike in crude oil prices to $145 per barrel in 2008 was driven by a combination of several factors that created a perfect storm for a rapid increase in oil prices:

1. Demand-Supply Imbalance:

High Global Demand: There was strong global demand for oil, especially from emerging economies like China and India, which were experiencing rapid industrialization and economic growth.
Limited Supply Growth: Oil production growth was constrained, particularly outside of OPEC (Organization of the Petroleum Exporting Countries), which limited the ability to meet increasing demand.

2. Geopolitical Tensions and Instability:

Middle East Tensions: Political tensions and instability in major oil-producing regions, particularly in the Middle East, raised concerns about potential disruptions to oil supply.
Iran and Iraq Concerns: Tensions surrounding Iran's nuclear program and ongoing conflict in Iraq added to supply concerns.

3. Speculative Trading and Investment:

Financial Speculation: Investment funds and speculative trading in commodity markets surged, driven by expectations of continued oil price increases and inflation hedging.
Weak Dollar: A weak US dollar made commodities, including oil, more attractive to investors seeking alternative investments.

4. Natural Disasters and Supply Disruptions:

Hurricanes and Natural Disasters: Events such as hurricanes in the Gulf of Mexico and other natural disasters temporarily disrupted oil production and refining capacities.

5. Market Dynamics and Futures Trading:

Futures Market Influence: The futures market, where traders speculate on future prices, played a significant role in amplifying price movements both upward and downward.

6. Global Economic Factors:

Strong Global Economic Growth: The period leading up to 2008 saw robust global economic growth, which increased demand for oil across various sectors, including transportation, manufacturing, and construction.
These factors combined to create a supply-demand imbalance and heightened market volatility, leading to the rapid escalation of crude oil prices to historic highs in mid-2008. However, this spike was followed by a sharp decline later in the year and into 2009, as the global financial crisis unfolded and economic activity slowed, significantly impacting oil demand.

```python
# Plotting the data through years

plt.figure(figsize = (12,6))
sns.lineplot(data = trade_my)
plt.title('Yearly Trade Value Trends for Malaysia (1995-2021)')
plt.xlabel('Year')
plt.ylabel('Trade Value ($)')

#plotting horizontal line for the average import and export
plt.axhline(y=avg_trade_balance, color="r", linestyle="-", linewidth=1, label=f"Average Import Trade Balance: ${avg_trade_balance:.2f}")


plt.legend(title='Action')
plt.grid(True)
plt.show()
```
