```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("medical_examination.csv")
df
```

```python
# Create the overweight column in the df variable
df["overweight"] = ((df["weight"] / ((df["height"]/100)**2)) >25).astype(int)
df
```

```python
# Normalize data by making 0 always good and 1 always bad. If the value of cholesterol or gluc is 1, set the value to 0. If the value is more than 1, set the value to 1.

df["cholesterol"] = df["cholesterol"].apply(lambda x:0 if x==1 else 1)
df["gluc"] = df["gluc"].apply(lambda x:1 if x==1 else 1)
df
```

```python
def draw_cat_plot():
    # Step 5: Create a DataFrame for the cat plot using pd.melt
    df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])

    # Step 6: Group and reformat the data in df_cat
    df_cat = df_cat.groupby(['cardio', 'variable', 'value'], as_index=False).size().rename(columns={'size': 'total'})

    # Step 7: Convert the data into long format and create the chart using sns.catplot
    fig = sns.catplot(x='variable', y='total', hue='value', col='cardio', data=df_cat, kind='bar').fig

    # Step 8: Return the figure for the output
    return fig
```

```python
draw_cat_plot()
```
